```python
import os
import csv
%matplotlib inline

import matplotlib.pyplot as plt


import collections
import pandas as pd
import numpy as np
import requests
import time
import datetime
```

* Loaded file 
* Readed  file and stored into Pandas data frame


```python
file_to_load = "Sacramento_Crime_Data_From_Current_Year.csv"

Sac_crime_df = pd.read_csv(file_to_load, delimiter=',', encoding="utf-8", dtype={'Time Occurred':str})
Sac_crime_df.head()
```

```python


# # format Occurence_date column and add "day of week" column
Sac_crime_df['Occurence_Date'] = Sac_crime_df['Occurence_Date'].str[:10]
Sac_crime_df.loc[:,'Occurence_Date'] = Sac_crime_df['Occurence_Date'].astype("datetime64[ns]")
Sac_crime_df['Day_of_Week'] = Sac_crime_df['Occurence_Date'].dt.weekday_name



Sac_crime_df.head()
```

```python
# # create "Days" column 
Sac_crime_df['Days'] = pd.DatetimeIndex(Sac_crime_df['Occurence_Date']).day

Sac_crime_df.head()
```

* Stored dataframe to a SQL database.

```python
from sqlalchemy import create_engine
engine = create_engine("postgresql://test:test@localhost:5432/ProjectTwo", echo=False)

Sac_crime_df.to_sql("sacramento_crime", con=engine,index=False, if_exists="replace")

# After running code set ObjectID as primary key
```
